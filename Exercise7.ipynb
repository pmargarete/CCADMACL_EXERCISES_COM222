```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.frequent_patterns import fpgrowth
```

```python
#reading data
url = "https://github.com/robitussin/CCADMACL_EXERCISES/blob/ff41fec98bd270f4bbeedb9b8849ae21fdeadfee/Groceries_dataset.csv?raw=true"
dataset = pd.read_csv(url)

dataset.head()
```

```python
client_basket = dataset.groupby(['Member_number', 'Date'])['itemDescription'].apply(lambda x: ', '.join(x))
client_basket
```

```python
client_basket.head()
```

```python
# Access the values of the Series directly instead of trying to access a column
client_basket = list(client_basket.apply(lambda x: x.split(",")))
client_basket
```

### 1. Apply one-hot encoding

```python
from mlxtend.preprocessing import TransactionEncoder

one_hot_transformer = TransactionEncoder()
basket_transform = one_hot_transformer.fit_transform(client_basket)
basket_transform
```

2. Use the Apriori Algorithm to find the most frequent sets

```python
client_basket = pd.DataFrame(basket_transform,columns=one_hot_transformer.columns_)
client_basket
```

3. Display the association rules using the lift metric

```python
client_basket = apriori(client_basket, min_support = 0.005, use_colnames = True)
client_basket.sort_values(['support'],ascending=False, inplace = True)
```

```python
client_basket_ar = association_rules(client_basket, metric="lift", min_threshold=1)

filtered_rules = client_basket_ar[(client_basket_ar['lift'] > 1) ]
display(filtered_rules)
```
