## Import the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import normalize
from sklearn.decomposition import PCA


```

## Load the data

```python
X = pd.read_csv('creditcard.csv')

# Dropping the CUST_ID column from the data
X = X.drop('CUST_ID', axis = 1)

# Handling the missing values
X.fillna(method ='ffill', inplace = True)

print(X.head())
```

## Pre-process the data

```python
# Scaling the data to bring all the attributes to a comparable level
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Normalizing the data so that
# the data approximately follows a Gaussian distribution
X_normalized = normalize(X_scaled)

# Converting the numpy array into a pandas DataFrame
X_normalized = pd.DataFrame(X_normalized)
```

## Reduce the dimensionality of the data using PCA

```python
pca = PCA(n_components = 2)
X_principal = pca.fit_transform(X_normalized)
X_principal = pd.DataFrame(X_principal)
X_principal.columns = ['P1', 'P2']
print(X_principal.head())
```

## Train a DBSCAN model using default values

```python
# Numpy array of all the cluster labels assigned to each data point
db_default = DBSCAN().fit(X_principal)
labels = db_default.labels_
```

## Plot the clustered data

```python
# Building the label to colour mapping
colours = {}
colours[0] = 'r'
colours[1] = 'g'
colours[2] = 'b'
colours[-1] = 'k'

# Building the colour vector for each data point
cvec = [colours[label] for label in labels]

# For the construction of the legend of the plot
r = plt.scatter(X_principal['P1'], X_principal['P2'], color ='r');
g = plt.scatter(X_principal['P1'], X_principal['P2'], color ='g');
b = plt.scatter(X_principal['P1'], X_principal['P2'], color ='b');
k = plt.scatter(X_principal['P1'], X_principal['P2'], color ='k');

# Plotting P1 on the X-Axis and P2 on the Y-Axis
# according to the colour vector defined
plt.figure(figsize =(9, 9))
plt.scatter(X_principal['P1'], X_principal['P2'], c = cvec)

# Building the legend
plt.legend((r, g, b, k), ('Label 0', 'Label 1', 'Label 2', 'Label -1'))

plt.show()
```

1. Train a DBSCAN model

```python
radius = 0.31797
core_point_samples = 1256

# Using the elbow method to find the optimal number of clusters
dbscan = DBSCAN(eps=radius, min_samples=core_point_samples)

# Fitting the model
model = dbscan.fit(X_principal)
```

```python
dbscan
```

```python
labels = model.labels_
```

```python
# Get number of clusters
n_clusters_ = len(set(labels)) #- (1 if -1 in labels else 0)
n_clusters
```

```python
# Number of Noisy data
n_noise_ = list(labels).count(-1)
n_noise_
```

```python
from sklearn import metrics

silhouette_score = metrics.silhouette_score(X_principal, labels)
```

```python
silhouette_score
```

```python
unique_labels = set(labels)
core_samples_mask = np.zeros_like(labels, dtype=bool)
core_samples_mask[dbscan.core_sample_indices_] = True

colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]
for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = [0, 0, 0, 1]

    class_member_mask = labels == k

    xy = X_principal[class_member_mask & core_samples_mask]
    # Convert xy to a NumPy array if it's not empty
    if not xy.empty:
        xy_array = xy.values # Convert the pandas DataFrame to a NumPy array
        plt.plot(
            xy_array[:, 0],  # Access data using NumPy array slicing
            xy_array[:, 1],
            "o",
            markerfacecolor=tuple(col),
            markeredgecolor="k",
            markersize=14,
        )

    xy = X_principal[class_member_mask & ~core_samples_mask]
    # Convert xy to a NumPy array if it's not empty
    if not xy.empty:
        xy_array = xy.values # Convert the pandas DataFrame to a NumPy array
        plt.plot(
            xy_array[:, 0],  # Access data using NumPy array slicing
            xy_array[:, 1],
            "o",
            markerfacecolor=tuple(col),
            markeredgecolor="k",
            markersize=6,
        )

plt.title(f"Estimated number of clusters: {n_clusters}")
plt.show()
```
